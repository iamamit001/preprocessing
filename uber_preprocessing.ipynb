```python
colsForOutliers=['Vehicles', 'Vehicles_normalized', 'visibility']
```

```python
def outliers_treatment(x):
    x=x.clip(upper=x.quantile(0.94))
    x=x.clip(lower=x.quantile(0.04))
    return x
```

```python
uber_data[colsForOutliers]=uber_data[colsForOutliers].apply(outliers_treatment)
```

```python
for i in colsForOutliers:
    sns.boxplot(x=uber_data[i])
    plt.show()
```
